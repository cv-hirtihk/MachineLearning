```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
#word2number is used to convert string like 'one' to equivalent number '1'
from word2number import w2n
import math
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/2_linear_reg_multivariate/Exercise/hiring.csv")
df.columns = ['experience', 'test_score', 'interview_score', 'salary']
df
```

```python
# Initally fill NaN using the meadian in the respective column
# We are using fillna() func to fill the specified String value for NaN
df.experience = df.experience.fillna("zero")
df
```

```python
# Convert String to their respective numbers
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
# To fill NaN in test_score column using the median
median_test_score = math.floor(df['test_score'].mean())
median_test_score
```

```python
df['test_score'] = df['test_score'].fillna(median_test_score)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop('salary', axis='columns'), df.salary)
```

```python
# co - efficient is the m 
coef = reg.coef_
coef
```

```python
# intercept is the C
intercept = reg.intercept_
intercept
```

```python
# x values are given as desired for prediction
coef[0] * 9 + coef[1] * 8.0 + coef[2] * 9 + intercept
```

```python
math.ceil(reg.predict([[9, 8.0, 9]]))
```
